From the dataset,fir a linear model on the relationship between the temperature in the living room in celcius(x=T2) and the temperature outside the building(y=T6).What is the Root Mean Squared Error(RMSE) in 3dp?

```python
#ASSIGNING X TO T2 AND Y TO T6
x=energy_df['T2']
y=energy_df['T6']


X=np.array(x)
Y=np.array(y)
```

```python
#arr=np.array(x)
#x_feature=arr.reshape(-1,1)
x_feature=X.reshape(-1,1)
```

```python

#SPLITING THE DATASET INTO TRAINING AND TESTING SETS
x_train,x_test,y_train,y_test=train_test_split(x_feature, Y,train_size=0.7,random_state=1)
#DEFINING THE MODEL
linear_m=LinearRegression()

#FITTING THE MODEL
linear_m.fit(x_train,y_train)

#PREDICTING THE MODEL
predicted_values=linear_m.predict(x_test)
```

```python
#RESIDUAL SUM OF SQUARES

rmse=np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```
