 EVALUATION METRICS FOR PERFORMANCE


1).MAE(MEAN ABSOLUTE ERROR)

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import pandas as pd

# Create a MinMaxScaler
scaler = MinMaxScaler()

# Normalize the dataset
normalized_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Drop the target columns 'Heating_Load' and 'Cooling_Load' from the features
features_df = normalized_df.drop(columns=['Heating_Load', 'Cooling_load'])
heating_target = normalized_df['Heating_Load']

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.3, random_state=1)

# Create a Linear Regression model
linear_model = LinearRegression()

# Fit the model to the training dataset
linear_model.fit(x_train, y_train)

# Obtain predictions
predicted_values = linear_model.predict(x_test)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,predicted_values)
round(r2_score,3)
```
